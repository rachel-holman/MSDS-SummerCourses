## Exercise 6.8



```python
df1 = pd.DataFrame({'name':['Rachel','Amy','Jess'],
                   'age':[22,18,30],
                   'favnum':[3,27,9]})
df2 = pd.DataFrame({'name':['Rachel','henry'],
                   'gender':['f','m'],
                   'idk':[1,2]})

#set indexes
df1 = df1.set_index('name')
df2 = df2.set_index('name')

pd.merge(df1, df2, on='name') #inner join with merge()
```

```python
df1.join(df2, how='outer') #outer join with join()
```
