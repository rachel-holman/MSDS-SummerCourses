## Congress API

```python
root = 'https://api.congress.gov/v3'
endpoint = '/member'
parameters = {'api_key': congresskey}
r = requests.get(root + endpoint, params= parameters,
                headers= headers)
r
```

```python
myjson = json.loads(r.text)
memberdf = pd.json_normalize(myjson, record_path=['members'])
memberdf
```

```python
bioguideId = memberdf.query("state=='Virginia' & district==5").reset_index()['bioguideId'][0]
bioguideId
```

```python
endpoint = f'/member/{bioguideId}/sponsored-legislation'
```

```python
endpoint
```

```python
r = requests.get(root + endpoint, params=parameters, headers= headers)
r
```

```python
myjson = json.loads(r.text)
bobgoodbills = pd.json_normalize(myjson, record_path=['sponsoredLegislation'])
```

```python
root= 'http://www.opensecrets.org/api/'
parameters = {'apikey': opensecretskey,
            'id': 'VA',
             'output': 'json',
            'method': 'getLegislators'}
r= requests.get(root, params = parameters, headers=headers)
r
```

```python
myjson = json.loads(r.text)
myjson = [x['@attributes'] for x in myjson['response']['legislator']]
valegislators= pd.json_normalize(myjson)
valegislators
```

```python
bobgoodid= valegislators.query('lastname=="Good"').reset_index()['cid'][0]
bobgoodid
```

```python
parameters = {'method': 'candContrib',
              'apikey': opensecretskey,
             'cid': bobgoodid,
             'output': 'json'}
r= requests.get(root, params = parameters, headers=headers)
r
```
